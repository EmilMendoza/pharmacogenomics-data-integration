# PharmGKB

Pharmacogenics Knowledge Base, or PharmGKB, is an onling repository of curated information about the complicated relationship between drug response and genetic makeup. This database includes several types of information on medication such as curated pathways, perscribing information, drug label annotations, clinical annotations and variant annotations. Relevant publications are also linked to the medication; which is where the information for annotations and pathways originate from.

The main purpose of PharmGKB is to readily offer accessible information to help facilitate personalized medicine. By understanging how a patient's individual genetic variation influences medication response, healthcare providers can make informed decisions on which medication to perscribe and ideal dosage. A personalized approach to medicine will lead to optimized treatment outcomes with minimized adverse reactions.

The PharmGKB website offers a user-friendly experience. In addition, PharmGKB also offers an Application Programming Interface, or API, with detailed documation on how to access and use the API. This allows the data too be programmatically accessed! The API provides endpoints for querying inforamtion about drugs, variant annotations, and more. For example, this API can be used to retreive information about a specific drug or to search for interactions.

# Creating a function to access the PharmGKB API

```python
import requests #importing the requests libary to access the API
import json
def PharmGKB(id,output): #creating a function that will create a query with an ID and return a file of the variants
    baseurl = 'https://api.pharmgkb.org/v1/data/variantAnnotation?relatedChemicals.accessionId=' #this is the base url for API for PharmGKB
    url = baseurl + id #by combining the baserul and pharmgkb id for our drug, we get an endpoint to use for the API
    response = requests.get(url) #this will get the associated variant annotations for the drug
    data = response.json() #creates a dict from the JSON encododed result
    with open(output, 'w') as file: #opening a file in write mode
        json.dump(data, file) #writes data to file in JSON format

PharmGKB('PA164712308','pharmGKBdata.json') #testing the function wtih ACE Inhibitors, plain, PharmGKB ID

#This function will create a take a PharmGKB drug id and return a file of associated variant annotations in json format into a file
```

# DGIdb

The Drug-Gene Interaction database, or the DGIdb, is another resource that provides inforamtion about interactions between variants and drugs. Similarly to PharmGKB, it is an aggregation of annotated and curated data. It also aims to provide easily accessible information to facilitate personalized medicine apporaches. DGIdb has similarities to PharmGKB, however DGIdb is a more focused database. It mainly focuses on how specific drugs interacts with genes and the mechanisms; meanwhile PharmGKB has a broader focus such as genetype-phenotype assocaitions and clinical guidelines.

# DGIdb API

The DGIdb API can be used to query for drug-gene interactions. This api uses GraphQL queries to access the API. Accessing data from the DGIdb API is very similar to PharmGKB, the difference only in the query. GraphQL queries has a schema that define the data that is being requested to be retrieved. DGIdb API documentation has information on how the query is set up.

```python
import requests #still using the requests library
import json #using json libary still
def DGIdb(drug, outputfile):
  query = """
{
  drugs(names: ["drug"]) {
    nodes {
      interactions {
        gene {
          name
          conceptId
          longName
        }
        interactionScore
        interactionTypes {
          type
          directionality
        }
        interactionAttributes {
          name
          value
        }
        publications {
          pmid
        }
        sources {
          sourceDbName
        }
      }
    }
  }
}"""

  url = 'https://dgidb.org/api/graphql'# The URL to the DGIdb GraphQL API endpoint

  # Make a request to the endpoint with the query
  response = requests.post(url, json={'query': query})
  data = response.json() #creates a dict from the JSON encododed result

  with open(outputfile, 'w') as file: #opening a file in write mode
      json.dump(data, file) #writes data to file in JSON format
```

##### Creating a function

```python
def extract_clinvar_variants(json_path):
    with open(json_path, 'r') as file:
        data = json.load(file)

    variants = data['result']
    uids = variants['uids']

    records = []
    for uid in uids:
        v = variants[uid]
        gene_symbol = v['genes'][0]['symbol'] if v.get('genes') else ''
        trait_set = v.get('germline_classification', {}).get('trait_set', [])
        trait_name = trait_set[0].get('trait_name', '') if trait_set else ''

        records.append({
            'uid': uid,
            'accession': v.get('accession'),
            'title': v.get('title'),
            'gene': gene_symbol,
            'protein_change': v.get('protein_change'),
            'variant_type': v.get('obj_type'),
            'molecular_consequence': ', '.join(v.get('molecular_consequence_list', [])),
            'classification': v.get('germline_classification', {}).get('description', ''),
            'trait': trait_name
        })

    return pd.DataFrame(records)
```

# All together now

ClinVar, PharmGKB, and DGIdb serves distinct and complementary purposes in the field of pharmacogenomics and personalized medicine. ClinVar focuses on the clinical significance of genetic variants, PharmGKB curates inforamtion on the reltionship between genetic varaintion and drug responses, and DGIdb provides information on the drug-gene interactions. Integrating data from these resoruces provides a hollistic understanding on how genetic variation influences drug therapy, guiding personalzied treatment startegies to minimize adverse reactions.

ClinVar does not provide direct information on how a genetic variant might affect drug response. PharmGKB focuses on drug responses but does not deeply cover disease pathogenicity of variants. DGIdb does not address variant-specific drug response. In total, a single database has it's strength but will not provide the full picture that can be gathered from combining data. ClinVar can identify whether the variant is medically significant whereas PharmGKB can provide explanations for how the variants can influence drug therapy. DGIdb can then determine where the gene can be pharmacologically targetted. This is a multi-layered approach and allows for researchs and clinicians to prioritize variants by filtering by both clinical significance and drug relevance, help guide therapy by matching patients to an optimal treatment plan based on their genomic profile and also mitigate risk by avoiding prescribing medidcations that could trigger adverse reactions on the patient's variants.
